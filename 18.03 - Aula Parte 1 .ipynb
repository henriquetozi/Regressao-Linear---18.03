```python
#Importando os dados
import pandas as pd

dados = pd.read_csv ("C:/Users/20221ENPRO0251/Downloads/Salary_Data.csv")

dados
```

```python
#Diagrama de Dispersão
import matplotlib.pyplot as plt

plt.scatter (dados['YearsExperience'], dados['Salary'], color = "lightcoral")
plt.title ("Salário vs Experiência")
plt.xlabel ("Anos de experiência")
plt.ylabel ("Salário (dólares americanos por ano)")
```

```python
#Separando X e Y

x = dados.iloc [: , :1]
y = dados.iloc [: , 1:]
```

```python
x
```

```python
y
```

```python
# Modelo 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

modelo = LinearRegression ()
modelo.fit (x,y)
```

```python
# Coefiecientes do modelo (betas)

print('Coeficiente angular (beta_1): ', modelo.coef_)
print ('Intercepto (beta_2): ', modelo.intercept_)
```

```python
# Gerando predições 

y_pred = modelo.predict (x)
```

```python
plt.scatter (x,y, color = "hotpink")
plt.plot (x,y_pred, color = "firebrick")
plt.title ("Salário vs Experiência")
plt.xlabel ("Anos de experiência")
plt.ylabel ("Salário anual em dólares americanos")
```

```python
# Coeficiente de determinação (R²)

r2_score (y, y_pred)
```
